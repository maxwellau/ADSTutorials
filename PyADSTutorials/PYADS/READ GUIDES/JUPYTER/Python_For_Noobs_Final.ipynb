````python
#Solution for Question 3, there are 2 solutions
#P.S. If you need more explaination just ask me :D

matrix = [[0,None,2,69],
          [3,4,5,69],
          [6,7,8,69],
          [9,10,11,69],
          [12,13,14,None]]

#row first search
def transpose_matrix_1(matrixInput):
  for i in range(len(matrixInput)):
    for j in range(len(matrixInput[i])):
        try:
          if matrixInput[i][j] != None:
            vars()[f"row{j}"].append(matrixInput[i][j])
          else:
            vars()[f"row{j}"].append(0)
        except:
          if matrixInput[i][j] != None:
            vars()[f"row{j}"] = [matrixInput[i][j]]
          else:
            vars()[f"row{j}"] = [0]
  transposed = []
  for i in range(len(matrixInput[0])):
    transposed.append(vars()[f"row{i}"])
    del vars()[f"row{i}"]
  return transposed


#column first search (reccomended)
def transpose_matrix_2(matrixInput):
  matrix_output = []
  for i in range(len(matrixInput[0])):
    row = []
    for j in range(len(matrixInput)):
      if matrixInput[j][i] != None:
        row.append(matrixInput[j][i])
      else:
        row.append(0)
    matrix_output.append(row)
  return matrix_output

solution_1 = transpose_matrix_1(matrix)
solution_2 = transpose_matrix_2(matrix)
print("Using first solution : ",solution_1)
print("Using first solution : ",solution_2)
print("Are they the same? ",solution_1 == solution_2)
````
